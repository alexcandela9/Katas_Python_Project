1. Escribe una funci贸n que reciba una cadena de texto como par谩metro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.

```python
# Definimos la funci贸n y decimos 
def contar_frecuencias(cadena):
    
    # Creamos un diccionario vac铆o donde guardaremos las frecuencias
    frecuencias = {}

    # Recorremos cada caracter de la cadena
    for letra in cadena:
        
        # Verificamos que el caracter no sea un espacio
        if letra != " ":
           
            # Si la letra ya existe, sumamos 1
            if letra in frecuencias:
                frecuencias[letra] += 1
           
            # Si no existe, la creamos con valor 1
            else:
                frecuencias[letra] = 1

    # Devolvemos el diccionario con las frecuencias de cada letra
    return frecuencias
```

```python
# Pedimos al usuario que escriba una cadena de texto
texto = input("Introduce una cadena de texto: ")

# Llamamos a la funci贸n y guardamos el resultado
resultado = contar_frecuencias(texto)

# Mostramos el diccionario con las frecuencias
print("Frecuencias de letras:")
print(resultado)
```
