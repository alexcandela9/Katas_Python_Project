10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una
excepción personalizada y maneja el error adecuadamente.

```python
class ListaVaciaError(Exception):              # Creamos una excepción personalizada
    pass                                       # No añadimos comportamiento extra

def promedio(lista):                           # Definimos la función
    if not lista:                              # Comprobamos si la lista está vacía
        raise ListaVaciaError("La lista está vacía")  # Lanzamos la excepción
    return sum(lista) / len(lista)             # Calculamos y devolvemos el promedio
```

```python
promedio([2,5,9])

promedio([])
```
