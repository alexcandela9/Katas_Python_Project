12. Genera una función que al recibir una frase devuelva una lista con la longitud de cada palabra. Usa la función map()

```python
def long_palabra(cadena):

    palabras = cadena.split()          # dividimos la frase por sus palabras

    return list(map(len, palabras ))   # devolvemos la longitud de cada una
```

```python
frase = "Hola mundo "

print(long_palabra(frase))
```
