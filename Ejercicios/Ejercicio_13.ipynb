13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en
mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()

```python
def Mayus_minus(caracteres):

    unicos = set(caracteres.lower())                                # obtenemos los elementos unicos de la cadena de caracteres

    return list(map(lambda x: (x.upper(), x.lower()), unicos))      # como queremos una tupla dentro de cada iterable de la lista ponemos los ()d a la operacion
```

Un set() es una colección no ordenada de elementos únicos e inmutables (hashables).

se distingue entre mayusculas y minusculas, por lo que ponemos un .lower() a caracteres.

```python
Mayus_minus("Holaaaah")
```
