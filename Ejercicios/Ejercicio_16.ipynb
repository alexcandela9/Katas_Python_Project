16. Escribe una función que tome una cadena de texto y un número entero n como parámetros y devuelva una lista de
todas las palabras que sean más largas que n. Usa la función filter()

```python
palabras_mas_largas = lambda texto, n: list(filter(lambda p: len(p) > n, texto.split()))
```

```python
frase = "hola mundo python"

frase.split()
```

```python
palabras_mas_largas(frase,4)
```

Sin lambda en la funcion:

```python
def palabras_mas_largas2(texto, n):
    return list(filter(lambda p: len(p) > n, texto.split()))
```

```python
palabras_mas_largas2(frase,4)
```
