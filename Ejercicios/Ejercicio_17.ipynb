17. Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2]
corresponde al número quinientos setenta y dos (572). Usa la función reduce()

```python
from functools import reduce                      # Importamos reduce de functools

def digitos_a_numero(digitos):                    # Definimos la función
    return reduce(lambda a, b: a * 10 + b, digitos)  # Construimos el número
```

reduce() es acumulativo recibiendo un solo valor final.

5*10 + 7 = 57

57*10 +2 = 572

```python
digitos = [5, 7, 2]

print(digitos_a_numero(digitos))
```

```python
digitos2 = [1, 3, 2, 1]

print(digitos_a_numero(digitos2))
```
