19. Crea una función lambda que filtre los números impares de una lista dada.

```python

filtrar_impares = lambda lista: list(filter(lambda x: x % 2 != 0, map(int, lista)))
```

Los numeros decimales los contaba como impares, para solucionarlo ponemos un int(x) convirtiendo el valor de entrada para la comprobacion a entero por truncacion, pero si era impar escribia el numero decimal, para solucionarlo quitamos el int() y pusimos un map() para transformar a entero toda la lista tanto para su comprobacion como para introducirlo en la lista si es impar.

```python
num = [1, 2, 3, 4.4, 7.5]

filtrar_impares(num)
```

Si ponemos un caracter no numerico, rompe, para que no rompa le podemos decir que solo acepte enteros:

```python
filtrar_impares_2 = lambda lista: list(filter(lambda x: isinstance(x, (int, float)) and x % 2 != 0, lista))

# A tener en cuenta que tambien va a contemplar como impares los numeros decimales.
```

```python
num = [1, 2, 3, 4.4, 7.5,"s"]

filtrar_impares_2(num)
```
