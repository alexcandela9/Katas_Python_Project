2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función map()

```python
# Creamos una lista de números
numeros = [1, 2, 3, 4, 5]

# Definimos una función que recibe un número y devuelve su doble
def duplicar(numero):
    return numero * 2

# Usamos map() para aplicar la función duplicar a cada elemento de la lista numeros
dobles = list(map(duplicar, numeros))

# Mostramos la lista con los valores duplicados
print(dobles)
```

La función map() sirve para aplicar una misma función a todos los elementos de una lista (o iterable), sin necesidad de usar bucles for.

Devuelve un objeto iterable, no una lista, por eso utilizamos list antes.

Podriamos dejarlo como iterable pero entonces no lo veriamos.

Tambien podemos hacerlo con lambda:

```python
numeros = [3, 9, 7, 5, 1]

# Usamos map() para aplicar una función a cada elemento de la lista y lambda
dobles = list(map(lambda x: x * 2, numeros))

# Mostramos la nueva lista con los valores duplicados
print(dobles)
```
