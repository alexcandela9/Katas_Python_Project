21. Crea una función que calcule el cubo de un número dado mediante una función lambda

```python
try:

    cubo = lambda x: x**3                                    # creamos la funcion cubo con su parametro de entrada y la operacion

    print(cubo (float(input("Introduce un numero"))))        # obligamos a que el dato introducido sea numero float

except ValueError:                                           # si no introduce un numero valido, muestro mensaje de error

    print("El numero no es correcto")
```

| Situación | Herramienta recomendada | Razón |
| :--- | :--- | :--- |
| **Entrada del usuario** | `try - except` | No puedes evitar que el usuario escriba letras en un campo numérico. |
| **Validar lógica propia** | `if - raise` | Si tu función espera un radio para un círculo, debes avisar si pasan un número negativo. |
| **Conexiones externas** | `try - except` | El servidor puede estar caído; es un error que "sucede" fuera de tu control. |
| **Límites de seguridad** | `if - raise` | Si un saldo de cuenta no puede ser menor a cero, detienes la operación tú mismo. |
