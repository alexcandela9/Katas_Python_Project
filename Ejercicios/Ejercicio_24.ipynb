24. Calcula la diferencia total en los valores de una lista. Usa la función reduce() .

```python
from functools import reduce                      # Importamos reduce

def diferencia_total(lista):                      # Definimos la función
    return reduce(lambda a, b: a - b, lista)      # Restamos los valores de forma acumulada
```

```python
num = [2, 3, 3]

diferencia_total(num)
```
