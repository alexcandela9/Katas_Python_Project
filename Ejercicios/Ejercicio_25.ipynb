25. Crea una función que cuente el número de caracteres en una cadena de texto dada.

```python
cuenta_car = lambda texto: len(texto)
```

```python
cuenta_car("Hola Python")
```

```python
cuenta_car([1, 2, 3, 8])
```

```python
cuenta_car(7)            # da error porque un objeto tipo entero no tiene longitud
```

```python
def cuenta_car2(texto):
    # Validamos que la entrada sea un string
    if not isinstance(texto, str):
        raise TypeError("La función cuenta_car2 solo acepta cadenas de texto (str).")
    
    return len(texto)
```

```python
cuenta_car2(7)          # ahora nos da el error programado
```

```python
try:
    print(cuenta_car("Hola")) # Bien
    print(cuenta_car(12345))  # Mal, pero la función lanza su propio mensaje
except TypeError as e:
    print(f"Aviso al usuario: {e}")
```
