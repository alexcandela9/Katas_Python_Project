26. Crea una función lambda que calcule el resto de la división entre dos números dados.

```python
def resto_division():                                   # Definimos la función
    try:
        a = int(input("Introduce el primer número: ")) # Pedimos el primer número
        b = int(input("Introduce el segundo número: "))# Pedimos el segundo número

        resto = (lambda x, y: x % y)(a, b)              # Lambda que calcula el resto
        print(f"El resto de la división es: {resto}")  # Mostramos el resultado

    except ValueError:                                  # Si el usuario introduce texto
        print("Error: debes introducir números enteros.")

    except ZeroDivisionError:                           # Si intenta dividir entre 0
        print("Error: no se puede dividir entre cero.")

```

```python
resto_division()
```
