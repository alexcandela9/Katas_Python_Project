27. Crea una función que calcule el promedio de una lista de números.

```python
def calcular_promedio():                                # Definimos la función
    try:
        cantidad = int(input("¿Cuántos números quieres introducir? "))

        if cantidad <= 0:                               # Validamos cantidad
            raise ValueError("La cantidad debe ser mayor que cero.")

        numeros = []                                    # Lista vacía para guardar números

        for i in range(cantidad):                       # Pedimos los números uno a uno
            numero = float(input(f"Número {i+1}: "))   # Permitimos decimales
            numeros.append(numero)                      # Añadimos a la lista

        promedio = sum(numeros) / len(numeros)          # Calculamos el promedio
        print(f"El promedio es: {promedio}")

    except ValueError as e:                              # Error de conversión o validación
        print(f"Error: {e}")
```

con except ValueError capturamos el error pero en este caso lo almacenamos en una variable llamada e (podria llamrse mensaje de error) y luego lo pasamos por pantalla porque si no hacemos esto desconocemos el mensaje de porque falla.

En nuestro error creado, al darle el tipo ValueError tambien captura el except, si le dieramos otro tipo no lo capturaria solo nos daria error y nos pasaria el mensaje.

```python
calcular_promedio()
```
