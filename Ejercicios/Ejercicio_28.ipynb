28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python
def primer_duplicado():                                 # Definimos la función
    try:
        cantidad = int(input("¿Cuántos elementos tendrá la lista? "))

        if cantidad <= 0:                               # Validamos cantidad
            raise ValueError("La lista debe tener al menos un elemento.")

        lista = []                                      # Lista para los valores

        for i in range(cantidad):                       # Pedimos los elementos
            valor = input(f"Elemento {i+1}: ")
            lista.append(valor)                         # Los añadimos a la lista

        vistos = set()                                  # Conjunto para controlar duplicados

        for elemento in lista:                          # Recorremos la lista
            if elemento in vistos:                      # Si ya apareció
                print(f"Primer elemento duplicado: {elemento}")
                return

            vistos.add(elemento)                        # Si no aparecio, guardamos el elemento y seguimos bucle

        print("No hay elementos duplicados.")           # Si termina el bucle

    except ValueError as mensaje_error:                              # Error en la entrada
        print(f"Error: {mensaje_error}")
```

```python
primer_duplicado()
```

set():

❌ No admite elementos repetidos

⚡ Es muy rápido para comprobar si algo existe

❌ No tiene posiciones ([0], [1], etc.)

```python
s = set()            # se ejecuta una sola vez, pero impone la no duplicidad permanente 
s.add("perro")
s.add("gato")
s.add("perro")

print(s)
```
