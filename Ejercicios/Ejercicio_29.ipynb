29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el
carácter '#', excepto los últimos cuatro.

```python
def enmascarar2(cadena):                            # Definimos la función diciendo que se le pasara una cadena
    texto = str(cadena)                            # Convertimos el valor a string
    
    return "#" * (len(texto) - 4) + texto[-4:]    # Enmascaramos excepto los últimos 4
```

```python
def enmascarar():                                                         # Definimos la función
    texto = input(str("Introduce una cadena"))                            # Pedimos un input y lo convertimos el valor a string
    visibles = min(4, len(texto))                                         # variable que coge el menor de ambos valores
    return "#" * (len(texto) - visibles) + texto[-visibles:]              # Enmascaramos excepto los últimos 4 sin len(texto)>4, sino excepto los ultimos del valor que tome len(texto) de esta forma nunca sera negativo.
```

La funcion esmascarar() y enmascarar2() hacen lo mismo solo que esmascarar() pide un input.

Aparte, solo se ha implementado la mejora de cambiar el valor 4 por una variable llamada visibles para que ninca se multiplique # por un numero negativo, aunque en este caso no cambia nada ya que aunque diera negativo no podia caracterizar # por un numero negativo.
