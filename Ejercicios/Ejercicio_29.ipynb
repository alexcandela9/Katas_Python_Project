29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el
carácter '#', excepto los últimos cuatro.

```python
def enmascarar2(cadena):                            # Definimos la función diciendo que se le pasara una cadena
    texto = str(cadena)                            # Convertimos el valor a string
    return "#" * (len(texto) - 1) + texto[-1:]    # Enmascaramos excepto los últimos 4
```

```python
def enmascarar():                                                         # Definimos la función
    texto = input(str("Introduce una cadena"))                            # Pedimos un input y lo convertimos el valor a string
    return "#" * (len(texto) - 4) + texto[-4:]                            # Enmascaramos excepto los últimos 4
```

```python
enmascarar()
```
