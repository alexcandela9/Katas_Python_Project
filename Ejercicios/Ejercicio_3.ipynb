3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

```python
# definimos la funcion con 2 parametros de entrada
def filtrar_palabras(lista_palabras, palabra_objetivo):
    
    # Creamos una lista vacía donde guardaremos las palabras que coincidan
    resultado = []
    
    # Recorremos cada palabra de la lista original
    for palabra in lista_palabras:
        
        # Comprobamos si la palabra objetivo está dentro de la palabra actual
        if palabra_objetivo in palabra:
            
            # Si está, añadimos la palabra a la lista resultado
            resultado.append(palabra) 
    
    # Devolvemos la lista con las palabras que cumplen la condición
    return resultado

```

```python
# Ejemplo de uso de la función
palabras = ["casa", "caseta", "perro", "cascada", "gato", "escaso"]

# Llamamos a la función pasando la lista y la palabra objetivo
resultado = filtrar_palabras(palabras, "cas")

# Mostramos el resultado por pantalla
print(resultado)
```

Otra forma mas rapida y limpia:

```python
def filtrar_palabras2(lista_palabras, palabra_objetivo):
    
    # Creamos y devolvemos directamente una lista con las palabras que contienen el objetivo, por eso lo metemos entre corchetes
    return [palabra for palabra in lista_palabras if palabra_objetivo in palabra]
```

```python
palabras = ["uno", "dos", "tres", "cuatro", "cinco"]

# Llamamos a la función pasando la lista y la palabra objetivo
resultado = filtrar_palabras2(palabras, "u")

# Mostramos el resultado por pantalla
print(resultado)
```
