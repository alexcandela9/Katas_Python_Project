30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
pero en diferente orden.

```python
def son_anagramas():                             # Definimos la función
    
    palabra1 = input(str("Introduce la primera palabra")).lower().replace(" ","")      # Pasamos a minuscula e ignoramos espacios
    palabra2 = input(str("Introduce la segunda palabra")).lower().replace(" ","")

    return sorted(palabra1) == sorted(palabra2)  # Comparamos letras ordenadas y devuelve True en caso de ser un anagrama o False si no lo es
```

sorted() ordena pero mantine la variable intacta, .sort() ordena pero modifica la variable a ese orden.

sorted() se puede usar en cualquier iterable (listas, tuplas, diccionarios, strings), .sort() solo en listas.

```python
son_anagramas()
```
