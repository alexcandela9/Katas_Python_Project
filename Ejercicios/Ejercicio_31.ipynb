31. Crea una función que solicite al usuario ingresar una lista de nombres y luego solicite un nombre para buscar en
esa lista. Si el nombre está en la lista, se imprime un mensaje indicando que fue encontrado, de lo contrario, se
lanza una excepción.

```python
lista = ["Alex", "Drako", "Byron"]

def buscar_nombre():
    try:
        nombre = str(input("Introduce un nombre: ")).title().replace(" ","")       # Pedimos el nombre al usuario

        if nombre in lista:                                                        # Comprobamos si está en la lista
            print("Nombre encontrado")                                             # Mostramos mensaje
        else:
            raise ValueError("Nombre no encontrado")                               # Lanzamos excepción


    except ValueError:
        print("Nombre no encontrado")
```
