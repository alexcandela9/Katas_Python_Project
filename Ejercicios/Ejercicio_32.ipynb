32. Crea una función que tome un nombre completo y una lista de empleados, busque el nombre completo en la lista y
devuelve el puesto del empleado si está en la lista, de lo contrario, devuelve un mensaje indicando que la persona
no trabaja aquí.

```python
def buscar_empleado(empleados):                               # La función solo recibe empleados, ya que podriamos poner el nombre como entrada pero como lo pedimos dentro no hace falta
    try:
        nombre = input("Introduce el nombre completo: ").title()  # Pedimos y normalizamos el nombre

        # CASO 1: empleados es un diccionario
        if isinstance(empleados, dict):                       # Si es un diccionario
            if nombre in empleados:                           # Buscamos la clave
                return f"Puesto: {empleados[nombre]}"         # Devolvemos el puesto
            else:
                return "La persona no trabaja aquí."

        # CASO 2: empleados es una lista de diccionarios
        elif isinstance(empleados, list):                     # Si es una lista
            for empleado in empleados:                        # Recorremos la lista
                if empleado["nombre"] == nombre:              # Comparamos nombres
                    return f"Puesto: {empleado['puesto']}"    # Devolvemos el puesto
            return "La persona no trabaja aquí."

        else:
            raise TypeError("Estructura de empleados no válida.")

    except KeyError:                                          # Si falta una clave
        return "Error: estructura de datos incorrecta."

    except TypeError as e:                                    # Error de tipo
        return f"Error: {e}"
```

```python
empleados_dict = {
    "Juan Perez": "Ingeniero",
    "Ana Lopez": "Diseñadora"
}

print(buscar_empleado(empleados_dict))
```

```python
empleados_lista = [
    {"nombre": "Juan Perez", "puesto": "Ingeniero"},
    {"nombre": "Ana Lopez", "puesto": "Diseñadora"}
]

print(buscar_empleado(empleados_lista))
```

```python
empleados_lista = (
    {"nombre": "Juan Perez", "puesto": "Ingeniero"},
    {"nombre": "Ana Lopez", "puesto": "Diseñadora"}
)

print(buscar_empleado(empleados_lista))
```
