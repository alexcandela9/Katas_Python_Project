33. Crea una función lambda que sume elementos correspondientes de dos listas dadas.

```python
def sumar_listas():

    try:                                                               # Arrancamos el bloque donde se puede producir algun error

        rango = int(input("Define el rango de las listas"))            # Pedimos el rango para ambas listas, obligando que sea un numero entero

        if rango <= 0:
            raise ValueError ("El rango debe ser superior a 0")        # Creamos nuestro error
        
        lista1 = []
        lista2 = []
        
        for i in range(rango):                                         # Con range le damos rangos al numero de veces que se ejecutara el bucle
            valor = float(input(f"Introduce el valor {i+1} de la lista 1"))    
            lista1.append(valor)                                       # Añadimos cada valor a lal lista

        for i in range(rango):
            valor = float(input(f"Introduce el valor {i+1} de la lista 2"))    
            lista2.append(valor)

        resultado = list(map(lambda x,y:x+y, lista1, lista2))       # En este caso al poner (lista1, lista2) map() cree que solo hay un agumento y lambda pide 2
    
        print(resultado)

    except ValueError as error:                                        # Capturamos el error, lo almacenamos y lo avisamos por pantalla
        print(f"el error es {error}")   
```

```python
sumar_listas()
```
