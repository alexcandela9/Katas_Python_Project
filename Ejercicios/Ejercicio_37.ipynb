Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras , reemplazar_palabras , eliminar_palabra. Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto .

Código a seguir:

    1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
    2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva. Tiene que devolver el texto con el remplazo de palabras.
    3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
    4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.

Caso de uso:

Comprueba el funcionamiento completo de la función procesar_texto

```python
def contar_palabras(texto):                                            # definimos la variable con su argumento
    conteo = {}                                                        # Creamos la variable diccionario vacia
    for palabra in texto.split():                                      # Dividimos la cadena y recorremos cada una con un bucle
        conteo[palabra] = conteo.get(palabra, 0) + 1                   # Añadimos al diccionario las palabras y las veces que se repiten
    return conteo

def reemplazar_palabras(texto, original, nueva):                       # Creamos una funcion con 3 argumentos
    return texto.replace(original, nueva)                              # Utilizamos replace para que busque la palabra a sustitur y la sustituya

def eliminar_palabra(texto, palabra):
    return texto.replace(palabra, "")       

def procesar_texto(texto, opcion, *args):                              # Creamos una funcion con 2 argumentos fijos y *args, que llama a las funciones anteriores
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        return reemplazar_palabras(texto, *args)
    elif opcion == "eliminar":
        return eliminar_palabra(texto, *args)
    else:
        raise ValueError("Opción no válida")                           # Lanzamos error si no elije ninguna de las opciones validas
```

```python
eliminar_palabra("holaaa hola", "hola")
```

```python
procesar_texto("holaaa hola","eliminar","hola")
```

```python
def contar_palabras2(texto):  
    # Creamos un diccionario vacío para guardar el conteo
    conteo = {}
    
    # Dividimos el texto en palabras y recorremos cada una
    for palabra in texto.split():
        # Si la palabra ya existe en el diccionario suma 1,
        # si no existe empieza en 0 y suma 1
        conteo[palabra] = conteo.get(palabra, 0) + 1
    
    return conteo


def reemplazar_palabra2(texto, original, nueva):
    palabras = texto.split(" ")  # Dividimos por espacios exactos
    
    resultado = []               # Creamos una lista vacia
    
    for p in palabras:
        if p == original:        # Solo si coincide exactamente
            resultado.append(nueva)    # Si la palabra de la iteracion de bucle es la original, añadimos la nueva palabra
        else:
            resultado.append(p)        # Si no lo es, añade la palabra de la iretacion
    
    return " ".join(resultado)   # Devuelve la union de cada palabra de la lista intercalando un espacio entre ellas




def eliminar_palabra2(texto, palabra):
    palabras = texto.split(" ")
    
    resultado = []
    
    for p in palabras:
        if p != palabra:
            resultado.append(p)
    
    return " ".join(resultado)




def procesar_texto2(texto, opcion, *args):
    # Según la opción elegida llamamos a la función correspondiente
    
    if opcion == "contar":
        return contar_palabras2(texto)
    
    elif opcion == "reemplazar":
        return reemplazar_palabra2(texto, *args)
    
    elif opcion == "eliminar":
        return eliminar_palabra2(texto, *args)
    
    else:
        raise ValueError("Opción no válida")
```
