38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.

```python
hora = int(input("Introduce la hora (0-23): "))     # Pedimos la hora como un entero.

if 6 <= hora < 12:                                  # Programamos los intervalos con if, elif y else.
    print("Es de día")
elif 12 <= hora < 20:
    print("Es de tarde")
else:
    print("Es de noche")
```
