39. Escribe un programa que determine qué calificación en texto tiene un alumno en base a su calificación numérica.
Las reglas de calificación son:
- 0 - 69 insuficiente
- 70 - 79 bien
- 80 - 89 muy bien
- 90 - 100 excelente

```python
def calificacion ():

    try:                                                                # Metemos todo en un try

        nota = float(input("Introduce tu calificacion numerica"))       # Pedimos la nota y tipo float

        if nota > 10 or nota < 0:
            raise ValueError("Introduce una nota valida")               # Lanzamos nuestro error para calificaciones no validas

        if nota < 7:                                                    # Programamos los intervalos y su repuesta por pantalla
            print("Insuficiente")
        elif nota >= 7 and nota <8:
            print("Bien")
        elif nota >= 8 and nota < 9:
            print("Muy Bien")
        else:
            print("Excelente")

    except ValueError as e:                                             # Mostramos el error por pantalla si lo hubiera
        print(f'Error:{e}')
```

```python
calificacion()
```

```python
# Otra forma:

nota = int(input("Introduce la calificación: "))

if 0 <= nota < 7:
    print("Insuficiente")
elif 7 <= nota < 8:
    print("Bien")
elif 8 <= nota <= 9:
    print("Muy bien")
elif 9 <= nota <= 10:
    print("Excelente")
else:
    print("Nota no válida")
```
