4. Genera una funci√≥n que calcule la diferencia entre los valores de dos listas. Usa la funci√≥n map()

```python
def diferencia_listas(lista1, lista2):                 # Definimos una funci√≥n que recibe dos listas
    return list(                                       # Convertimos el resultado de map en una lista
        map(                                           # Usamos map para aplicar una funci√≥n a los elementos
            lambda a, b: a - b,                        # Funci√≥n lambda que resta b a a
            lista1,                                    # Primera lista que se recorrer√° elemento a elemento
            lista2                                     # Segunda lista que se recorrer√° en paralelo
        )
    )
```

üîπ ¬øQu√© hace map() aqu√≠?

map() aplica una funci√≥n a la vez sobre los elementos de varias listas.

En este caso:

Primera iteraci√≥n ‚Üí a = lista1[0], b = lista2[0]

Segunda iteraci√≥n ‚Üí a = lista1[1], b = lista2[1]

Y as√≠ sucesivamente‚Ä¶

```python
lista1 = [5, 10, 15]                   # Definimos la primera lista
lista2 = [1, 2, 3]                     # Definimos la segunda lista

resultado = diferencia_listas(lista1, lista2)  # Llamamos a la funci√≥n
print(resultado)                        # Mostramos el resultado
```

Otra forma con for sin map():

En este caso si la longitud de la lista 2 es mas corta, rompe.

```python
def diferencia_listas2(lista1, lista2):            # Definimos una funci√≥n que recibe dos listas
    resultado = []                                # Creamos una lista vac√≠a para guardar las diferencias

    for i in range(len(lista1)):                  # Recorremos indices, no elementos, de un rango equivalente a la longitud de la lista
        diferencia = lista1[i] - lista2[i]        # Ese indice i da valor a ambas listas y resta sus elementos de su misma posicion
        resultado.append(diferencia)              # A√±adimos la diferencia a la lista resultado

    return resultado                              # Devolvemos la lista con todas las diferencias
```

```python
lista1 = [5, 10, 15]                   # Definimos la primera lista
lista2 = [1, 2, 3]                     # Definimos la segunda lista

resultado = diferencia_listas2(lista1, lista2)  # Llamamos a la funci√≥n
print(resultado)                        # Mostramos el resultado
```

Si lista2 es mas corta, rompe, solucion:

```python
def diferencia_listas2_2(lista1, lista2):                # Definimos la funci√≥n
    resultado = []                                    # Lista para resultados
    longitud_max = max(len(lista1), len(lista2))     # Longitud de la lista m√°s larga

    for i in range(longitud_max):                     # Recorremos todas las posiciones
        valor1 = lista1[i] if i < len(lista1) else 0 # Tomamos valor o 0 si no existe
        valor2 = lista2[i] if i < len(lista2) else 0 # Tomamos valor o 0 si no existe
        resultado.append(valor1 - valor2)             # Restamos y guardamos

    return resultado                                  # Devolvemos la lista
```

```python
lista1 = [5, 10, 15,22]                   # Definimos la primera lista
lista2 = [1, 2, 3]                     # Definimos la segunda lista

resultado = diferencia_listas2_2(lista1, lista2)  # Llamamos a la funci√≥n
print(resultado)                        # Mostramos el resultado
```

Usando zip():

En este caso si tienen diferente longitud, el bucle para a la iteracion de la lista mas corta.

```python
def diferencia_listas3(lista1, lista2):        # Definimos la funci√≥n
    resultado = []                            # Lista donde guardaremos las diferencias

    for a, b in zip(lista1, lista2):          # Recorremos ambas listas a la vez
        resultado.append(a - b)               # Restamos y a√±adimos el resultado a la lista

    return resultado                          # Devolvemos la lista final
```

```python
lista1 = [5, 10, 15]                   # Definimos la primera lista
lista2 = [1, 2, 3]                     # Definimos la segunda lista

resultado = diferencia_listas3(lista1, lista2)  # Llamamos a la funci√≥n
print(resultado) 
```

zip() se encarga de emparejar posiciones.

Pero si quisieramos que si tienen una longitud diferente no pierda informacion:

```python
from itertools import zip_longest                 # Importamos zip_longest

def diferencia_listas3_3(lista1, lista2):            # Definimos la funci√≥n
    resultado = []                                # Lista para guardar las diferencias

    for a, b in zip_longest(lista1, lista2, fillvalue=0):  # Recorremos ambas listas
        diferencia = a - b                        # Restamos los valores (0 si no existe)
        resultado.append(diferencia)              # Guardamos el resultado

    return resultado                              # Devolvemos la lista final
```

```python
lista1 = [10, 20, 30, 40]
lista2 = [1, 2]

resultado = diferencia_listas3_3(lista1, lista2)
print(resultado)
```
