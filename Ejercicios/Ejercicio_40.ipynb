40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o
"triangulo") y datos (una tupla con los datos necesarios para calcular el área de la figura).

```python
import math                                             # Importamos la libreria math

def area(figura, datos):
    if figura == "rectangulo":
        base, altura = datos                            # desenpaquetamos los datos
        return base * altura
    elif figura == "circulo":
        (radio,) = datos
        return math.pi * radio ** 2
    elif figura == "triangulo":
        base, altura = datos
        return (base * altura) / 2
    else:
        raise ValueError("Figura no válida")            # Lanzamos error si no es una de las figuras
```

Cuando paso varios valores, Python necesita saber cuál es cuál:

Con *args podemos poner los argumentos necesarios esperar una cantidad fija.

Pero como en este caso necesitas en concreto una cantidad exacta, no nos valdria otra diferente. Aun asi lo haremos con *args:

```python
import math

def area(figura, *args):                                                # Le decimos que no hay una cantidad fija de argumentos
    figura = figura.lower()                                             # Normalizamos la entrada a minusculas de la figura
    
    if figura == "rectangulo":
        if len(args) != 2:
            raise ValueError("El rectángulo necesita base y altura")    # Lanzamos error si nos da una cantidad diferente de argumentos a los necesarios
        base, altura = args
        return base * altura
    
    elif figura == "circulo":
        if len(args) != 1:
            raise ValueError("El círculo necesita solo el radio")
        radio = args[0]
        return math.pi * radio ** 2
    
    elif figura == "triangulo":
        if len(args) != 2:
            raise ValueError("El triángulo necesita base y altura")
        base, altura = args
        return (base * altura) / 2
    
    else:
        raise ValueError("Figura no válida")
```
