5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver una tupla que contenga la media y el estado.

```python
def calcular_media_estado(notas, nota_aprobado=5):   # Definimos la función con un parámetro opcional
    media = sum(notas) / len(notas)                  # Calculamos la media sumando y dividiendo
    estado = "aprobado" if media >= nota_aprobado else "suspenso"  # Determinamos el estado
    return media, estado                              # Devolvemos una tupla con la media y el estado
```

```python
notas = (9, 1, 8, 1, 2)

print(calcular_media_estado(notas))
```

```python
notas = [9, 1, 8, 1, 2]

print(calcular_media_estado(notas, 4))
```

Al llamar a la funcion por defecto considera 5 el corte de aprobado si no decimos nada, si le damos un valor nuevo, este cambia.
