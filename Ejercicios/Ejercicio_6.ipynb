6. Escribe una función que calcule el factorial de un número de manera recursiva.

```python
def factorial(n):                     # Definimos la función factorial
    if n == 0 or n == 1:              # Caso base de la recursividad
        return 1                      # El factorial de 0 o 1 es 1
    return n * factorial(n - 1)       # Llamada recursiva
```

Al hacer una llamada recursiva hay que establecer un caso base, ya que si no nunca pararia.

La funcion va multiplicando n * factorial (n-1)

Como el factorial (n-1) = (n-1) * factorial de ((n-1)-1)

Va multiplicando hasta encontrarse con el caso de 1 o 0 que cierran el ciclo.

```python
factorial(4)
```

Con for:

```python
def factorial_2(n):
    
    resultado = 1

    for i in range(1, n+1):

        resultado *= i

    return resultado
```

```python
factorial_2(4)
```
