7. Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map()

```python
def tuplas_a_strings(lista_tuplas):           # Definimos la función
    return list(                              # Convertimos el resultado en lista
        map(str, lista_tuplas)                # Convertimos cada tupla en string
    )
```

La funcion str convierte cualquier objeto a una cadena de texto.

```python
lista = [(1,2), (3,4), (5,6)]
print(tuplas_a_strings(lista))
```

```python
tupla = (1,2)
print(tuplas_a_strings(tupla))
```

Con for:

```python
def tuplas_a_strings_2(lista_tuplas):
    
    resultado = []
    
    for i in lista_tuplas:

        resultado.append(str(i))

    return resultado    


```

```python
lista = [(1,2), (3,4), (5,6)]
print(tuplas_a_strings_2(lista))
```
