8. Escribe un programa que pida al usuario dos números e intente dividirlos. Si el usuario ingresa un valor no numérico
o intenta dividir por cero, maneja esas excepciones de manera adecuada. Asegúrate de mostrar un mensaje
indicando si la división fue exitosa o no.

```python
try:
    num1 = float(input("Introduce el primer número: "))   # Pedimos el primer número
    num2 = float(input("Introduce el segundo número: "))  # Pedimos el segundo número
    resultado = num1 / num2                                # Intentamos la división
    print("La división fue exitosa:", resultado)           # Mostramos el resultado
except ValueError:
    print("Error: debes introducir números válidos.")     # Error si no es numérico
except ZeroDivisionError:
    print("Error: no se puede dividir entre cero.")       # Error división por cero
```

### Cómo funciona paso a paso
 
1. Se entra en el bloque `try`.
2. Python empieza a ejecutar las instrucciones dentro.
3. Si no aparece ninguna excepción, se ignoran los `except`.
4. Si aparece una excepción, Python:
   - busca el `except` correspondiente al tipo de error,
   - ejecuta solo ese `except`,
   - no sigue ejecutando el resto del bloque `try`.
5. Luego puede ejecutarse `else` o `finally` según estén presentes.
