9. Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista
excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre",
"Serpiente Pitón", "Cocodrilo", "Oso"].Usa la función filter()

```python
def filtrar_mascotas(mascotas):                                # Definimos la función
    
    prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]  # Lista prohibida
    
    return list(filter(lambda m: m.title() not in prohibidas, mascotas))            # Convertimos a lista y le damos formato de titulo

```

```python
lista = ["mapache", "perro", "Cocodrilo", "oSO"]

print(filtrar_mascotas(lista))
```

Con for():

```python

def filtrar_mascotas_2(mascotas):
    
    resultado = []
    prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
    
    for mascota in mascotas:
        
        if mascota.title() not in prohibidas:
            
            resultado.append(mascota) 
    
    return resultado

```

```python
lista = ["mapache", "perro", "Cocodrilo", "oSO"]

print(filtrar_mascotas_2(lista))
```
